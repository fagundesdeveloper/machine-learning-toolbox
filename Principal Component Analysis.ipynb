### Principal Component Analysis

The purpose of principal component analysis is identify patterns to reduce the dimensions of the dataset with minimal loss of information. 

PCA uses an orthogonal transofrmaiton - a form of linear transformation 


PCA is a form of linear transformation, related to Multiple Discriminant Analysis(MDA).# perhaps more in a later part about mda and pca and the differences between them. Pca is often followed by mda

In PCA were are ignoring class labels and projecting the entire set of data onto a different subspace to deistinguish between patterns that belong to different classes.

```python
import numpy as np

mu_vec1 = np.array([0,0,0]) #sample means
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]]) #covariance matrix
class1_sample = np.random.multivariate_normal(mu_vec1, cov_mat1, 20).T

class1_sample
```
