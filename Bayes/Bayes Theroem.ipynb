# Bayes Theroem

$$ P(A|B) = \frac{P(B|A) * P(A)}{P(B)} $$


Example: 
Given two machines 1 and 2, machine 1 produces 30 wrenches and hour and machine 2 produces 20 wrenches an hour. In total 1% of parts are defective and if we were to pick out a part from the defective pile there is a 50% chance that it is from machine 1 and a 50% chance that it is from machine 2. What is the chance that a part from machine 2 being defective. 

Thus: 
- P(Mach2)=(20/50) = .4 
- P(Defect)=.01
- P(Mach2|Defect) =.5
- P(Defect|Mach2) = ?


$$ P(Defect|Mach2)= \frac{P(Mach2|Defect) * P(Defect)}{P(Mach2)} $$

$$ P(Defect|Mach2)= \frac{.5 * 0.01}{0.4}  = 0.0125 $$
