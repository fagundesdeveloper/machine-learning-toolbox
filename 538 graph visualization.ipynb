```python
import numpy as np
import pandas as pd
from numpy.random import randn

link = 'http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv'
women_majors = pd.read_csv(link)
print(women_majors.info())
```

```python
%matplotlib inline
under_20 = women_majors.loc[0, women_majors.loc[0] < 20]
under_20

under_20_graph = women_majors.plot(x = 'Year', y = under_20.index, figsize = (12,8))
print('Type:', type(under_20_graph))
```
