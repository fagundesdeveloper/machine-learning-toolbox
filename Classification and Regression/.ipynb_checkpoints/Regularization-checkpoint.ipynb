# Regularization

Regularization is a way to avoid overfitting by penalizing the magnitude of regression coefficients along with minimizing the error between predicted and actual observations. In simple terms, it reduces paramters and simplifies the model. Regularization adds penalties to more complex models and then sorts potential models from least overfit to greatest. 

**C** is the parameter for regularization. Small values of C constrain the model more. Here is a picture showing the sparsity of coefficients when applying l1 and l2 penalty with different values of c.



Based on a stack overflow post sklearn uses l2 regularization with a constant of 1 by default, but these can be set manually using the 'penalty' and 'c' flags. 

## L1/ Lasso regularization

An acronym for least absolute shrinkage and selection operator. Lasso regularization adds a penalty equivalent to *absolute value of the magnitude* of coefficients.

L1 tends to constrain coefficients more than L2 and this helps it perform feature selection in sparse feature spaces. In practice, however l2 is going to provide better results.

## L2/ Ridge regularization 

L2 adds a penalty equivalent to *square of the magnitude* of coefficients.

According to the LIBLINEAR FAQ they have not found a practical example where l1 beats l2. You should likelly always use l2 in practice, even in a situation where you use L1 for feature slection- you should use l2 on the remaining variables for best results. 

## Elastic Net

Elastic net is often the preferred solution to lasso and ridge regression because it solves the limitations of both models. Given strong prior knowledge it could make sense to choose lasso or ridge but in the abscense of knowledge elastic net makes the most sense. Using the elastic net requires computing an additional parameter, which is it's major limitation. 

Simply, the elastic net is the convex sum of the ridge and lasso regularizations. It's quadratic penalty makes the loss funciton strictly convex and it therefore has a unique minimum. 

GLMNET is the standard algorithm for estimating elastic net regularization solutions.
