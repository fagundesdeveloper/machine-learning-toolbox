```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.jointplot(x='total_bill', y ='tip', data=tips, kind='reg')
```

```python
sns.pairplot(tips, hue="sex", palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical Plots

```python
import numpy as np
```

```python
sns.barplot(x= 'sex', y= 'total_bill', data= tips, estimator= np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split='true')
```

```python
sns.stripplot(x = 'day', y= 'total_bill', data=tips, hue='sex', split= True, jitter=True)
```

```python
sns.swarmplot(x = 'day', y='total_bill', data= tips, color='black')
sns.violinplot(x= 'day', y='total_bill', data= tips)
```

# Matrix Plots

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
# use heatmap to look at the covariance matrix 
tc = tips.corr()
sns.heatmap(tc, annot= True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month', columns="year", values="passengers")
```

```python
sns.heatmap(fp, cmap='magma', linecolor= 'black', linewidths= 1)
```

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale= 1)
```

# Regression Plots

```python
sns.lmplot(x='total_bill', y='tip', hue='sex', markers=['o', 'v'], 
           scatter_kws={'s': 100}, data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', col='sex', row='time', 
           aspect=0.6, size=8, data=tips)
```

# Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
#g.map(plt.scatter)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker') 
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker') 
g.map(plt.scatter, 'total_bill', 'tip')
```
